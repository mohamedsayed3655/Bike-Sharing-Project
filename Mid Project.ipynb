# Describe Each Columns In The Dataset 

| اسم العمود       | الشرح                                                                 |
| ---------------- | --------------------------------------------------------------------- |
| **datetime**     | التاريخ والوقت لكل ساعة (Timestamp) الخاص بعدد رحلات الدراجات.        |
| **count**        | إجمالي عدد رحلات الدراجات في تلك الساعة (الهدف / Target).             |
| **holiday**      | 1 إذا كان اليوم عطلة رسمية – 0 إذا كان يوم عادي.                      |
| **workingday**   | 1 إذا كان يوم عمل (ليس عطلة ولا ويك إند)، 0 إذا كان إجازة أو ويك إند. |
| **temp**         | درجة الحرارة الفعلية (بوحدة °C).                                      |
| **feels_like**   | درجة الحرارة المحسوسة (بالاعتماد على الرياح والرطوبة).                |
| **temp_min**     | أقل درجة حرارة متوقعة خلال الساعة.                                    |
| **temp_max**     | أعلى درجة حرارة متوقعة خلال الساعة.                                   |
| **pressure**     | الضغط الجوي بوحدة hPa (hectopascal).                                  |
| **humidity**     | نسبة الرطوبة المئوية (%) في الهواء.                                   |
| **wind_speed**   | سرعة الرياح (بوحدة المتر/الثانية).                                    |
| **wind_deg**     | اتجاه الرياح (بالدرجات من 0 إلى 360).                                 |
| **rain_1h**      | كمية الأمطار (ملم) خلال آخر ساعة.                                     |
| **snow_1h**      | كمية الثلج (ملم) خلال آخر ساعة.                                       |
| **clouds_all**   | نسبة الغيوم في السماء (%) من 0 إلى 100.                               |
| **weather_main** | وصف مختصر لحالة الطقس (مثل Clear – Clouds – Rain – Snow).             |

# Feature Engineering

```python
df['hour']=df['datetime'].dt.hour
df['day']=df.datetime.dt.day_name()
df['month']=df.datetime.dt.month
df['year']=df.datetime.dt.year
```

```python
print(df.temp.min())
print(df.temp.max())
```

```python
def wether_temp(x):
    if x>=-12.79 and x<=5:
        return 'Cold'
    elif x>=6 and x<=15:
        return 'Cool'
    elif x>=16 and x<=25:
        return 'Warm'
    else:
        return 'Hot'
df['weather_temp']=df.temp.apply(wether_temp)    
```

```python
df
```

```python
def Today_times(h):
    if h>=0 and h<=5:
        return "Night"
    elif h>=6 and h<=11:
        return "Morning"
    elif h>=12 and h<=16:
        return "Noon"
    else:
        return "Afternoon"
df["Today_times"] = df["hour"].apply(Today_times)
```

```python
df
```

```python
def season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

df["season"] = df["month"].apply(season)
```

```python
df.humidity
```

```python
df['heat_feeling']=((df.humidity * df.temp)/100).abs()
```
